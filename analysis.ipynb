### Cell count cutoff

The cell counts associated to various relaxation/decay times

```python
# Counts the number of cells with various relaxation times
# Used to generate minor subplot of Figure 3 C

paramtable = pd.read_csv('Data/finalparameters.csv')

tau_thresh = np.arange(0, 2.625, 0.125)
cell_counts = np.zeros_like(tau_thresh)

for itr, thresh in enumerate(tau_thresh):
    unit_list_indexes = np.where(paramtable.iloc[:, 3].values >= thresh)[0]
    unit_list = paramtable.iloc[unit_list_indexes, 0].values
    cell_counts[itr] = len(unit_list)

plt.figure(4)
plt.plot(tau_thresh, cell_counts, '.k-', markersize=18)
plt.xlabel('Relaxation Time Threshold (s)')
plt.ylabel('Number of Cells')
plt.show()
```
